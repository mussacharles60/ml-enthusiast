```python
import os
# import requests
import http.client
from urllib.parse import urlparse, urljoin
from bs4 import BeautifulSoup

# HTML file containing the links
html_file_path = 'data/useful_swahili_words.html'

# URL of the webpage
base_url = 'https://www.glcom.com/hassan/lessons/useful_swahili_words.html'

# Directory to save downloaded MP3 files
output_directory = 'data/downloaded_mp3s'

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)
```

```python
# Custom headers to be added to the requests
# custom_headers = {
#     'Accept': '*/*',
#     'Accept-Encoding': 'identity;q=1, *;q=0',
#     'Accept-Language': 'en-US,en;q=0.9',
#     'Dnt': '1',
#     'Host': 'learn101.org',
#     'Proxy-Connection': 'keep-alive',
#     'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36'
# }
# custom_headers_template = {
#     ':authority': 'www.glcom.com',
#     ':method': 'GET',
#     ':path': '/hassan/audio/dialogue/nitakula.MP3',
#     ':scheme': 'https',
#     'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
#     'Accept-Encoding': 'gzip, deflate, br',
#     'Accept-Language': 'en-US,en-TZ;q=0.9,en;q=0.8,sw-TZ;q=0.7,sw;q=0.6',
#     'Cache-Control': 'no-cache',
#     'Dnt': '1',
#     'Pragma': 'no-cache',
#     'Sec-Ch-Ua': '"Google Chrome";v="117", "Not;A=Brand";v="8", "Chromium";v="117"',
#     'Sec-Ch-Ua-Mobile': '?0',
#     'Sec-Ch-Ua-Platform': '"Windows"',
#     'Sec-Fetch-Dest': 'document',
#     'Sec-Fetch-Mode': 'navigate',
#     'Sec-Fetch-Site': 'none',
#     'Sec-Fetch-User': '?1',
#     'Upgrade-Insecure-Requests': '1',
#     'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36'
# }

# # Function to download an MP3 file from a URL
# def download_mp3(mp3_url, output_directory):
#     try:
#         # Create a copy of the custom headers template
#         custom_headers = custom_headers_template.copy()
        
#         # Set the dynamic :path based on the mp3_url
#         custom_headers[':path'] = urlparse(mp3_url).path
        
#         response = requests.get(mp3_url, headers=custom_headers, stream=True)
#         if response.status_code == 200 or response.status_code == 206:
#             # Extract the filename from the URL
#             url_path = urlparse(mp3_url).path
#             file_name = os.path.join(output_directory, os.path.basename(url_path))

#             # Save the MP3 file locally with the same name as in the URL
#             with open(file_name, 'wb') as mp3_file:
#                 for chunk in response.iter_content(chunk_size=1024):
#                     if chunk:
#                         mp3_file.write(chunk)
#             print(f'Downloaded: {file_name}')
#         else:
#             print(f'Failed to download: {mp3_url}, Status Code: {response.status_code}')
#     except Exception as e:
#         print(f'Error downloading: {mp3_url}, Error: {str(e)}')


def download_mp3(mp3_url, output_directory):
    try:
        # Parse the URL to extract hostname, path, and port
        parsed_url = urlparse(mp3_url)
        hostname = parsed_url.hostname
        path = parsed_url.path
        port = parsed_url.port if parsed_url.port else 443  # Default to HTTPS port

        # Establish a connection to the server
        connection = http.client.HTTPSConnection(hostname, port=port)
        
        # Construct the request headers with custom headers
        headers = {
            # ':authority': 'www.glcom.com',
            # ':method': 'GET',
            # ':path': path,
            # ':scheme': 'https',
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
            'Accept-Encoding': 'gzip, deflate, br',
            'Accept-Language': 'en-US,en-TZ;q=0.9,en;q=0.8,sw-TZ;q=0.7,sw;q=0.6',
            'Cache-Control': 'no-cache',
            'Dnt': '1',
            'Pragma': 'no-cache',
            'Sec-Ch-Ua': '"Google Chrome";v="117", "Not;A=Brand";v="8", "Chromium";v="117"',
            'Sec-Ch-Ua-Mobile': '?0',
            'Sec-Ch-Ua-Platform': '"Windows"',
            'Sec-Fetch-Dest': 'document',
            'Sec-Fetch-Mode': 'navigate',
            'Sec-Fetch-Site': 'none',
            'Sec-Fetch-User': '?1',
            'Upgrade-Insecure-Requests': '1',
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36'
        }

        # Send the HTTP GET request with custom headers
        connection.request('GET', path, headers=headers)
        response = connection.getresponse()

        if response.status == 200 or response.status == 206:
            # Extract the filename from the URL
            url_path = urlparse(mp3_url).path
            file_name = os.path.join(output_directory, os.path.basename(url_path))

            # Save the MP3 file locally with the same name as in the URL
            with open(file_name, 'wb') as mp3_file:
                while True:
                    chunk = response.read(1024)
                    if not chunk:
                        break
                    mp3_file.write(chunk)
            print(f'Downloaded: {file_name}')
        else:
            print(f'Failed to download: {mp3_url}, Status Code: {response.status}')
    except Exception as e:
        print(f'Error downloading: {mp3_url}, Error: {str(e)}')
```

```python
# Load the HTML file
with open(html_file_path, 'r', encoding='utf-8') as html_file:
    html_content = html_file.read()

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find all MP3 links in the HTML content
# mp3_links = [link['href'] for link in soup.find_all('a', href=True) if link['href'].endswith('.mp3')]
mp3_links = [urljoin(base_url, link['href']) for link in soup.find_all('a', href=True) if link['href'].lower().endswith(('.mp3', '.mp3'))]


mp3_links
```

```python
# Download each MP3 file
for mp3_link in mp3_links:
    mp3_url = mp3_link
    download_mp3(mp3_url, output_directory)

print('Download complete.')
```
