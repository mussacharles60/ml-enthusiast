```python
import os
import shutil

# Define the source directory where files are located
source_directory = 'data/clips'

# Define the destination directory where alphabet folders will be created
destination_directory = 'data/alphabet_folders'

# Create the destination directory if it doesn't exist
os.makedirs(destination_directory, exist_ok=True)

# Function to create alphabet folders and move files
def organize_files(source_dir, dest_dir):
    # Loop through files in the source directory
    for filename in os.listdir(source_dir):
        if os.path.isfile(os.path.join(source_dir, filename)):
            # Get the first character of the filename (assumes filenames start with letters)
            first_letter = filename[0].upper()

            # Create the letter folder in the destination directory if it doesn't exist
            letter_folder = os.path.join(dest_dir, first_letter)
            os.makedirs(letter_folder, exist_ok=True)

            # Move the file to the corresponding letter folder
            source_file = os.path.join(source_dir, filename)
            dest_file = os.path.join(letter_folder, filename)

            # Check if the destination file already exists and avoid overwriting
            if not os.path.exists(dest_file):
                shutil.copy(source_file, dest_file)
                print(f'Moved: {filename} to {letter_folder}')
            else:
                print(f'Skipped: {filename} (File already exists in {letter_folder})')

# Call the function to organize the files
organize_files(source_directory, destination_directory)

print('Organizing complete.')
```
